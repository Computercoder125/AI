Purpose: To simulate a solution of the Water Jug Puzzle step by step

```python

#This class contains all the methods to perform actions on jugs such as emptying and updating their amounts

class WaterJugPuzzle():

  def __init__(self, capacity, amount):
    self.capacity = capacity;
    self.amount = amount;

  #Check for invalid states (guard against them)

  def is_jug_full(self):
    if(self.capacity == self.amount): return True
    return False

  def is_jug_empty(self):
    if(self.amount == 0): return True
    return False

  def is_valid_state(self):
      return 0 <= self.amount <= self.capacity

  #Set amounts

  def update_amount(self, newAmount):
      if self.is_valid_state():
          self.amount += newAmount

      elif (self.is_jug_full()):
        raise ValueError("Jug is full!")

      else:
        raise ValueError("Jug is empty!")

  #save the new amount of jugs if valid
  def save_state(self, amount):
    if(self.is_valid_state):
      self.amount = amount

  #Print each state as it goes through the rules

  def print_state(self):
      print("Amount in jug: " f"{self.amount}")

#Methods for following Rules:

  #Fill Jugs

  def fill_jug(self):

      self.save_state(self.amount)

      self.amount = self.capacity

  #Empty Jugs

  def empty_jug(self):

      self.save_state(self.amount)

      if(not(self.is_jug_empty())):
        self.amount = 0

```

```python
#iterating through the solution

left_jug = WaterJugPuzzle(5, 0) #left_jug has a capacity of 5 and an amount of 0

right_jug = WaterJugPuzzle(3, 0) #right jug has a capaicty of 3 and initial amount of 0

print("INITIAL STATE: ")

left_jug.print_state()

right_jug.print_state()

print("-----------------------")
#Rule 1: Fill Left Jug

left_jug.fill_jug()

left_jug.print_state()

right_jug.print_state()

print("-----------------------")
#Rule 2: Fill Right Jug with water from Left Jug

right_jug.fill_jug()

left_jug.update_amount(-3)

left_jug.print_state()

right_jug.print_state()

print("-----------------------")
#Rule 3: Empty Small Jug

right_jug.empty_jug()

right_jug.save_state(right_jug.amount)

left_jug.print_state()

right_jug.print_state()


print("-----------------------")
#Rule 4: Pour the remaining two litres from the large jug into the small jug

left_jug.empty_jug()

right_jug.update_amount(2)

left_jug.print_state()

right_jug.print_state()

print("-----------------------")
#Rule 5: Refill the large jug

left_jug.fill_jug()

left_jug.print_state()

right_jug.print_state()

print("-----------------------")
#Rule 6: Pour one litre from the large jug into the small jug

left_jug.update_amount(-1)

right_jug.update_amount(1)

#Now, print out the final states for both jugs

print("Final Amount in large jug: " f"{left_jug.amount}" "\n" "Final Amount in small jug: "
      f"{right_jug.amount}")
```
