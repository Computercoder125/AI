**(h)** Use Python code to calculate the proportion of vehicles in the data set that have
combined mpg that is within 1.5 standard deviations of the mean. (Use the sample standard
deviation.) Is the result consistent with what is predicted by Chebyshev's theorem?

```python
ssd = mpg.std(ddof=1)
mean = mpg.mean()
proportion = len(mpg[(mpg >= mean - 1.5*ssd) & (mpg <= mean + 1.5*ssd)])/len(mpg)
proportion
```

I think this result is pretty consistent. According to Chebyshev's theorem, 1-(1/(1.5)^2) = 1 - 0.444... = about 0.5556, meaning at least 55.56 percent of observations lie within 1.5 SD of the mean. According to the calucations above, the proportion was 92.5 percent, and since 92.5 > 55.56, this result indeed matches Chebyshev's theory.

**(c)** Make a deep copy of the DataFrame in (a) that will be used for imputing on
missing values. On the DataFrame copy, impute on the missing values for each quantitative
variable using the median. Then, impute on the missing values for each categorical
variable using the mode.

```python
df_categorical = CIA["net_users"]

df_num = CIA[["gdp", "oil_prod", "educ", "roadways"]]

CIACopy = CIA.copy()

CIACopy[["gdp", "oil_prod", "educ", "roadways"]] = CIACopy[["gdp", "oil_prod", "educ", "roadways"]].fillna(df_num.median())

CIACopy["net_users"] = CIACopy["net_users"].fillna(df_categorical.mode()[0])

CIACopy
```

Verify that the imputed DataFrame has no more missing values by counting the total
number of missing values in the imputed DataFrame.

```python
CIACopy.isnull().sum().sum() #since 0 is listed to the right of every column title, there are no missing values
```

**(d)** Create a scatterplot of gdp vs educ where the points are colored by the net_users
variable and the size of the point is indicated by the roadways variable. Order the
categories of net_users from smallest to largest in the legend for the plot by specifying
*category_orders* in the px.scatter() function. Also, set the width and height of the
scatterplot to 800 and 600, respectively.

```python
gdp = CIACopy["gdp"]
educ = CIACopy["educ"]
net_users = CIACopy["net_users"]
roadways = CIACopy["roadways"]

fig = px.scatter(
    CIACopy,
    x="educ",
    y="gdp",
    color="net_users",
    size="roadways",
    category_orders= {"net_users": [">0%", ">5%", ">15%", ">35%", ">60%"]},
    width=800,
    height=600
)

fig.show()
```
