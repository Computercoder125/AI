I again used a modified version of BFS that explores neighboring cells in order of increasing effort. Instead of tracking the total cost, each step records the maximum absolute difference in heights along the path so far. This BFS algorithm continues expanding until it reaches the bottom-right cell, always relaxing edges when a smaller maximum effort is found. This ensures the smallest possible maximum height difference of all paths can be found.
